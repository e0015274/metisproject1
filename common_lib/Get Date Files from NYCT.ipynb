```python
import pandas as pd
import numpy as np
import datetime
```

```python
def Get_Data_Files(start_date='190706', num_of_files=3):
    """
    Pull files from NYCT site and merge them.
    URL: http://web.mta.info/developers/turnstile.html
    ---
    df = Get_Data_Files(start_date, num_of_files)
        start_date: The first date of the newest data file you wish to download (format = "yymmdd")
        num_of_files: Number of data files to download and merge dated by start_date, 
                      then all older files in previous weeks 
    """
    Next_Date = lambda date: (datetime.datetime.strptime(date, "%y%m%d") \
                              - datetime.timedelta(days=7)).strftime("%y%m%d")
    Get_FileName = lambda date: "http://web.mta.info/developers/data/nyct/turnstile/turnstile_" \
                              + date + ".txt"
    
    df = pd.DataFrame()
    req_date = start_date
    for i in range(num_of_files):
        print(">> Download and merging this:", Get_FileName(req_date))
        df2 = pd.read_csv(Get_FileName(req_date))
        df = df.append(df2, ignore_index=True)
        req_date=Next_Date(req_date)
    return df
```
